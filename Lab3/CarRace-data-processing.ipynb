Notebook to process data of Car Race experiment and make plots.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ev = np.load('../logs/evaluations.npz')
# ev2 = np.load('./logs/evaluations_2M.npz')
print(ev)
# print(ev2)
x1 = ev['timesteps']
# x2 = ev2['timesteps'] + int(1e6)
# x = np.concatenate((x1, x2))
print(x1)
mean_res1 = [np.mean(ev['results'][i]) for i in range(ev['results'].shape[0])]
std_res1 = [np.std(ev['results'][i]) for i in range(ev['results'].shape[0])]
# mean_res2 = [np.mean(ev2['results'][i]) for i in range(ev2['results'].shape[0])]
# std_res2 = [np.std(ev2['results'][i]) for i in range(ev2['results'].shape[0])]
# mean_res = np.concatenate((mean_res1, mean_res2))
# std_res = np.concatenate((std_res1, std_res2))
print(mean_res1, std_res1)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x1, mean_res1, label='Mean Reward')
plt.fill_between(x1, np.array(mean_res1) - np.array(std_res1), np.array(mean_res1) + np.array(std_res1), color='#0066FF', alpha=0.3, label='1 std')
plt.xlabel('Timesteps')
plt.ylabel('Reward')
plt.title('Mean and Variance of Rewards')
plt.legend()
plt.grid(True)
plt.savefig('reward_mean_variance.png')
plt.show()
```
